**2. Membuat statistik dari data setiap atribut yang bertype numerik**

Tipe data numerik merupakan tipe data yang mewakili data atau nilai berjenis bilangan, makanya juga dapat disebut tipe data bilangan. Tipe data numerik dibagi atas 3 jenis yaitu integer, float, dan complex. 

**2.1 Statistik median data TV**

Untuk menghitung median sebenarnya sudah ada library nya sendiri yakni median(),  disini akan menghitung median atau nilai tengah dari data TV yang ada. Rumus menghitung median adalah  
Rumus Median dengan Jumlah data ganjil :

$$M_{e} = \frac{X (n + 1)}{2}$$

Rumus Median dengan Jumlah Data Genap :

$$M_{e} = \frac{X_\frac{n}{2} + X\left (\frac{n}{2}+1 \right )}{2}$$

```python
dataset['tv'].median()
```

**2.2 Statistik median data Radio**

```python
dataset['radio'].median()
```

**2.3 Statistik mean data TV**

Mean atau rata - rata dihasilkan dari pembagian antara jumlah nilai keseluruhan dengan banyaknya data yang diolah. Mean disimbolkan dengan xÂ¯ rumusnya adalah 

$$mean = \frac{x_{1}+x_{2}+x_{3} ... x_{n}}{n}$$



```python
dataset['tv'].mean()
```

**2.4 Statistik mean data Radio**

```python
dataset['radio'].mean()
```

**2.5 Statistik modus data TV**

Modus adalah data yang paling sering muncul. Modus merupakan ukuran pemusatan untuk menyatakan fenomena yang paling banyak terjadi. Sekumpulan data yang diperoleh, memungkinkan untuk memiliki nilai modus yang tidak tunggal atau mungkin juga tidak memilikinya.

```python
dataset['tv'].mode()
```

**2.6 Statistik modus data radio**

**3. Mendeskripsikan kemiringan/skwened dari data setiap variabel**

Kemiringan atau kecondongan (skewned) adalah tingkat ketidak simetrisan atau kejauhan simetri dari sebuah distribusi. Skewned diartikan sebagai kemiringan distribusi data. 

**3.1 Skwened TV**

```python
data_tv.skew() 
```

```python
plt.figure()
sns.distplot(data_tv)
plt.show()
```

**3.2 Skwened Radio**

```python
data_radio.skew()
```

```python
plt.figure()
sns.distplot(data_radio)
plt.show()
```
